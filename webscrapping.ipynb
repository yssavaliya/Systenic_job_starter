```python
import requests
from bs4 import BeautifulSoup
import smtplib
import smtplib, ssl
import webbrowser

#credentials
port = 587  # For starttls
smtp_server = "smtp.gmail.com"
sender_email = input("Enter Your Gmail: ")
receiver_email = input("Enter Reciver G-Mail: ")
password = input("Enter Password: ")

# Making a GET request
getting_request = requests.get('https://www.systenics.com/jobs/')

# Parsing the HTML
html = BeautifulSoup(getting_request.content, 'html.parser')

#printing company
title = html.find('title')
print(title.text)
print("\n")

#getting the header
content0 = html.find('h3')
print(content0.text)

# description,type,location
content1 = html.find('ul',{'class': 'details' })
print(content1.text)


message ="""\
Subject: {title}
{content0}
{content1}
""".format(title = title.text,content0 = content0.text, content1=content1.text)

context = ssl.create_default_context()
with smtplib.SMTP(smtp_server, port) as server:
    server.ehlo() 
    server.starttls(context=context)
    server.ehlo() 
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, message) 
```
